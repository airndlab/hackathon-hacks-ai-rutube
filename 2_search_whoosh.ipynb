```python
from whoosh.fields import Schema, TEXT, ID
from whoosh.index import create_in
from whoosh.filedb.filestore import RamStorage
from whoosh.analysis import RegexTokenizer, LowercaseFilter, Filter
from whoosh.support.charset import accent_map
from whoosh.index import open_dir

from nltk.stem.snowball import RussianStemmer

# Ensure NLTK Snowball data is downloaded
import nltk
nltk.download('snowball_data')

# Create a filter using NLTK's Russian stemmer
class RussianStemFilter(Filter):
    def __init__(self):
        self.stemmer = RussianStemmer()

    def __call__(self, tokens):
        for t in tokens:
            t.text = self.stemmer.stem(t.text)
            yield t

# Create a custom analyzer for Russian | RussianStemFilter()
russian_analyzer = RegexTokenizer() | LowercaseFilter() 
# Define the schema for your video data
schema = Schema(
    video_id=ID(stored=True, unique=True),
    title=TEXT(stored=True, analyzer=russian_analyzer)
)

# Create an in-memory index
storage = RamStorage()
index = storage.create_index(schema)

# Choose a directory to store the index
#indexdir = "our_indexes/whoosh3"
#index = create_in(indexdir, schema)
#index = open_dir(indexdir)
```

```python
import math

batch_size = 10000  # Adjust this based on your system's capability

# Calculate total batches
total_batches = math.ceil(len(filtered_df) / batch_size)

for batch_num in range(total_batches):
    with index.writer() as writer:
        # Calculate start and end index for each batch
        start_index = batch_num * batch_size
        end_index = start_index + batch_size

        # Convert the specific chunk of DataFrame to dictionaries
        for row in filtered_df[start_index:end_index].to_dicts():
            writer.add_document(
                video_id=str(row['video_id']),
                title=row['video_title']
            )

    # Print or log after each batch
    print(f"Batch {batch_num + 1}/{total_batches} indexed.")

print("Indexing complete.")
```

```python
from whoosh.qparser import MultifieldParser

def search_videos(index, query_str, limit=5):
    with index.searcher() as searcher:
        
        query = MultifieldParser(["title"], schema=index.schema).parse(query_str)
        results = searcher.search(query,limit=limit)
        return [(hit['video_id']) for hit in results]
```

```python
# Search for "search engine"
search_query = "мужское"
search_results = search_videos(index, search_query)
print(len(search_results))
# Print out the search results
for result in search_results:
    print(result)
```
